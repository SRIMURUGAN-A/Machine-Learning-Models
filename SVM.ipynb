#svm algorithum



```python
import numpy as np
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
from sklearn import svm
```

```python
x, y = make_blobs(n_samples = 40, centers = 2, random_state =20)
model = svm.SVC(kernel='linear', C=1000)
model.fit(x,y)
```

```python
plt.scatter(x[:, 0], x[:, 1], c=y, s=30, cmap=plt.cm.Paired)
plt.title('svm')  

# Get coefficients for hyperplane
w = model.coef_[0]  # coefficients of the hyperplane
intercept = model.intercept_[0]  # intercept
a = -w[0] / w[1]  # slope of the hyperplane

# Generate x values for the decision boundary
xx = np.linspace(min(x[:, 0]), max(x[:, 0]))
yy = a * xx - intercept / w[1]  # calculate y values for the decision boundary

# Plot the decision boundary
plt.plot(xx, yy, 'k-', label='Decision Boundary')

# Plot margin lines
margin = 1 / np.sqrt(np.sum(w ** 2))
yy_margin_up = yy + margin
yy_margin_down = yy - margin
plt.plot(xx, yy_margin_up, 'k--', label='Margin')
plt.plot(xx, yy_margin_down, 'k--')

# Add title and legend
plt.title('SVM with Hyperplane and Margins')
plt.legend()
plt.show()
```

```python
import numpy as np
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
from sklearn import svm

# Generate data
x, y = make_blobs(n_samples=40, centers=2, random_state=20)

# Train SVM model
model = svm.SVC(kernel='linear', C=1)
model.fit(x, y)

# Plot data points
plt.scatter(x[:, 0], x[:, 1], c=y, s=30, cmap=plt.cm.Paired)

# Plot support vectors
plt.scatter(
    model.support_vectors_[:, 0],
    model.support_vectors_[:, 1],
    s=100,
    facecolors='none',
    edgecolors='k',
    label='Support Vectors',
)

# Get coefficients for hyperplane
w = model.coef_[0]  # coefficients of the hyperplane
intercept = model.intercept_[0]  # intercept
a = -w[0] / w[1]  # slope of the hyperplane

# Generate x values for the decision boundary
xx = np.linspace(min(x[:, 0]), max(x[:, 0]))
yy = a * xx - intercept / w[1]  # calculate y values for the decision boundary

# Plot the decision boundary
plt.plot(xx, yy, 'k-', label='Decision Boundary')

# Plot margin lines
margin = 1 / np.sqrt(np.sum(w ** 2))
yy_margin_up = yy + margin
yy_margin_down = yy - margin
plt.plot(xx, yy_margin_up, 'k--', label='Margin')
plt.plot(xx, yy_margin_down, 'k--')

# Add title and legend
plt.title('SVM with Hyperplane and Margins')
plt.legend()
plt.show()
```
