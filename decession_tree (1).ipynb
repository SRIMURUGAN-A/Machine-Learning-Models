```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn import tree
```

```python
import pandas as pd
df = pd.read_csv('data_dt.csv', index_col=None)
df.head()
```

```python
df = pd.DataFrame(df)
df.head()
```

```python
df.isnull().sum()
```

```python
y =df['RepaymentStatus']
x = df.drop('RepaymentStatus', axis=1)
x.shape
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3,random_state =100)
```

```python
random_classifier = DecisionTreeClassifier(random_state =100, criterion = "entropy", max_depth= 3, min_samples_leaf=5)
random_classifier.fit(x_train, y_train)
```

```python
pred = random_classifier.predict(x_test)
pred
```

```python
score  =  accuracy_score(y_test, pred)
print(y_test.shape, pred.shape)
score
```

```python
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree

# Plot the decision tree
plt.figure(figsize=(12,8))
plot_tree(random_classifier, filled=True, feature_names=x.columns, class_names=y.unique(), fontsize=10)
plt.show()
```
